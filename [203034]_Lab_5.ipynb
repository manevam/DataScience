```
from transformers import AutoTokenizer, XLMForSequenceClassification,RobertaForSequenceClassification,AutoModelForTokenClassification
tokenizer = AutoTokenizer.from_pretrained('xlm-roberta-base')
```

```
model = XLMForSequenceClassification.from_pretrained('xlm-roberta-base',num_labels =3)
```

```
train = train.drop(['id', 'keyword','location'], axis=1)
train.head(5)
```

```
test = res.drop(['id', 'keyword','location'], axis=1)
test.head(5)
```

```
train1 = train.rename({'target': 'label'}, axis=1)
train1.head()
```

```
test1 = test.rename({'target': 'label'}, axis=1)
test1.head()
```

```
train1.to_csv('trainThis.csv',index=False)
```

```
test1.to_csv('testThis.csv',index=False)
```

```
from datasets import load_dataset
dataset = load_dataset("csv", data_files={"train": "trainThis.csv", "test": "testThis.csv"})
dataset
```

```
encoded_dataset = dataset.map(lambda t: tokenizer(t['text'],  truncation=True), batched=True,load_from_cache_file=False)
```

```
encoded_dataset['test']
```

```
encoded_dataset['train']
```

```
from transformers import TrainingArguments, Trainer
```

```
arg = TrainingArguments(
    "sentiment",
    learning_rate=5e-5,
    num_train_epochs=3,
    per_device_eval_batch_size=12,
    per_device_train_batch_size=12,
    seed=19
)
```

```
trainer = Trainer(
    model=model,
    args=arg,
    tokenizer=tokenizer,
    train_dataset=encoded_dataset['train'],
    eval_dataset=encoded_dataset['test']
)
```

```
trainer.train()
```

```
y_pred = trainer.predict(encoded_dataset['test'])
```

```
y_pred = y_pred.predictions
```

```
import numpy as np
y_pred = [np.argmax(y_pred[i]) for i in range(0,len(y_pred))]
y_pred
```

```
test1.head()
```

```
from sklearn.metrics import classification_report,f1_score
f1_score(test1['label'],y_pred)
```

```
print(classification_report(test1['label'],y_pred))
```

```
#@title Load the Universal Sentence Encoder's TF Hub module
from absl import logging

import tensorflow as tf

import tensorflow_hub as hub
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import re
import seaborn as sns

module_url = "https://tfhub.dev/google/universal-sentence-encoder/4" #@param ["https://tfhub.dev/google/universal-sentence-encoder/4", "https://tfhub.dev/google/universal-sentence-encoder-large/5"]
model = hub.load(module_url)
print ("module %s loaded" % module_url)
def embed(input):
  return model(input)
```

```
logging.set_verbosity(logging.ERROR)
embeddings = embed(test1.text.values)
embeddings
```

```
from sklearn.cluster import KMeans

kM = KMeans(n_clusters=3)
kM.fit(embeddings)

clusters = kM.labels_.tolist()
```

```
from sklearn.decomposition import PCA
import numpy as np

data = embeddings
pca = PCA(2)
 
data1 = pca.fit_transform(data)
```

```
kmeans = KMeans(n_clusters=3)
 
label = kmeans.fit_predict(data1)

u_labels = np.unique(label)
 
labels = kmeans.labels_
```

```
import matplotlib.pyplot as plt
import seaborn as sns

for i in u_labels:
    plt.scatter(data1[label == i , 0] , data1[label == i , 1] , label = i)
    
plt.legend()
plt.show()
```
